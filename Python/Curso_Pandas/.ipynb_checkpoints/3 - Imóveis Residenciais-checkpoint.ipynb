## Imóveis Residenciais

```python
#importando a biblioteca numpy
import pandas as pd
#Importando a biblioteca seaborn
import seaborn as sns
```

```python
#Criação de um objeto DataFrame
dados = pd.read_csv("Dados/aluguel.csv", sep =";")
dados.head()
```

```python
#listagem dos tipos existentes na base de dados
todos_tipos = list(dados["Tipo"].drop_duplicates())

#Listagem criada manualmente com os tipo "residenciais"
Residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
Residencial
```

```python
#Metodo para comparação do dataFrame separando o que é "residencial"
selecao = dados["Tipo"].isin(Residencial)
selecao.head(5)
```

```python
#Novo DataFrame somente com os dados de imóveis Residenciais
dados_residencial = dados[selecao]
dados_residencial
```

```python
dados_residencial = dados_residencial.sort_values(by="Tipo")
dados_residencial
```

```python
dados_residencial.index = range(dados_residencial.shape[0])
dados_residencial
```

#### Exportando a base de dados

```python
dados_residencial.to_csv("Dados/aluguel_residencial.csv", sep=";", index = False)
```

#### Comparação dos tipos de Imóveis Residenciais

```python
#Contagem por tipo de dados
count_imoveis = dados_residencial["Tipo"].value_counts()
count_imoveis
```

```python
#Visualização gráfica da quantidade de cada tipo de imóvel residencial
sns.catplot(x="Tipo", kind = "count", data = dados_residencial)
```

```python
#Notou-se que existem muitos apartamentos em comparação com os outros tipos. Quanto?
total_imoveis = count_imoveis.sum()
total_ape = count_imoveis.loc["Apartamento"]
total_outros = total_imoveis - total_ape

print("Total de imoveis residenciais:",total_imoveis)
print("Total de imóveis do tipo apartamentos:",total_ape)
print("Total de imóveis do tipo !apartamentos:",total_outros)
```

```python
#Exibição de forma gráfica a comparação entre apartamentos e outros Tipos
dados = {
    "Tipo":["Apartamentos", "Outros"],
    "Total":[total_ape, total_outros]
}
sns.barplot(x="Tipo", y="Total", data =dados)
```
